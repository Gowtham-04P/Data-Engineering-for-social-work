## Keep a item-price column with the original price of the item.

```python
def clean_item_price(price: str):  

    if ' each est.'in price:
        price = price.replace(' each est.','') 
    if type(price)!=str:
        pass
    elif '£' in price:
        price = price.replace('£', '')
    elif 'p' in price:
        price = price.replace('p', '')
        price = '0.' + price

    if '-' in price:
       price = price.split('-')[0]

     
    return float(price)
df['price-fixed']=df['price'].astype(str).apply(clean_item_price)
```

```python
selected_rows = df['price'].astype(str).str.contains('-')

df.loc[selected_rows,'size'] = df.loc[selected_rows,'size'].str.split('-').str[0].str.replace('(','')
```

```python
df['price']=df['price-fixed']
df.drop(columns=['price-fixed'],inplace=True)
```

```python
bananas= df[
    df['name'].str.contains('Bananas', case=False, na=False) & 
     (df['categories'].str.contains('everyday', case=False, na=False) | 
    ~df['categories'].str.contains('everyday', case=False, na=False))
]
```

```python
with pd.option_context('display.max_rows', None,):
    print(bananas)
```

```python
bananas_remove=["Ella's Kitchen Peaches & Bananas","Ella's Kitchen Puree Bananas","Ella's Kitchen Blueberries Apples", "Ella's Kitchen Bananas & Apples", "Ella's Kitchen Bananas & Coconuts", "Innocent Strawberries & Bananas Fruit Smoothie", "Innocent Strawberries & Bananas Fruit Smoothie", "Ella's Kitchen Bananas & Apples", "Innocent Pineapples, Bananas & Coconuts Fruit Smoothie","Innocent Strawberries & Bananas Fruit Smoothie", "Innocent Strawberries & Bananas Fruit Smoothie", "Innocent Pineapples, Bananas & Coconuts Fruit Smoothie", "Ella's Kitchen Peaches & Bananas", "Ella's Kitchen Bananas & Apples", "Innocent Strawberries & Bananas Fruit Smoothie", "Innocent Strawberries & Bananas Fruit Smoothie", "Innocent Pineapples, Bananas & Coconuts Fruit Smoothie"]
bananas= bananas[~bananas['name'].str.contains('|'.join(bananas_remove), case=False, na=False)]
bananas
```

```python
bananas_plot = bananas[['id','name','size','price','categories']]
bananas_plot=bananas_plot.sort_values('price')
bananas_plot
```

```python
(
    ggplot(bananas_plot, aes(x='name', y='price')) +
    geom_line(color='red', size=1.2) +
    labs(
        title='Price of Bananas',
        x='Product Name',
        y='Price ($)'
    ) +  
    theme(
        axis_text_x=element_text(hjust=1),  
        plot_title=element_text(size=14, face='bold')  
    ) + ggsize(600, 800)+
        theme_grey() +\
        flavor_high_contrast_dark()      
)
```
